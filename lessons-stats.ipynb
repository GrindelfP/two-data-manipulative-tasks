```python
import numpy as np
import pandas as pd
```

```python
curriculum = pd.read_csv('data/curriculum.csv', sep=';')

curriculum
```

```python
curriculum = curriculum.fillna(0)

curriculum
```

```python
curriculum_matrix = curriculum.to_numpy()
curriculum_matrix
```

```python
def sum_of_row_with_step(row, step, start_index):
    sum = 0
    for i in range(start_index, len(row), step):
        sum += row[i]

    return sum

truncated_matrix = np.zeros((14, 8), dtype=object)

for i in range(truncated_matrix.shape[0]):
    for j in range(8):
        if j == 0:
            truncated_matrix[i, j] = curriculum_matrix[i, j]
        else:
            truncated_matrix[i, j] = sum_of_row_with_step(curriculum_matrix[i], 7, j)

truncated_matrix
```

```python
combined_curriculum = []
headers = curriculum.columns[:8].to_list()

combined_curriculum.append(headers)

for i in range(truncated_matrix.shape[0]):
    combined_curriculum.append(truncated_matrix[i])

combined_curriculum
```

```python
output = []
for i in range(1, len(combined_curriculum)):
    output.append(f'{combined_curriculum[i][0]}: {combined_curriculum[i][1]} часов всего. Из них: \n')
    output.append(f'{combined_curriculum[i][2]} {combined_curriculum[0][2]}. ')
    output.append(f'{combined_curriculum[i][3]} {combined_curriculum[0][3]}. ')
    output.append(f'{combined_curriculum[i][4]} {combined_curriculum[0][4]}. ')
    output.append(f'{combined_curriculum[i][5]} {combined_curriculum[0][5]}. ')
    output.append(f'{combined_curriculum[i][6]} {combined_curriculum[0][6]}. ')
    output.append('\n\n')

for line in output:
    if not line.startswith('0'):
        print(line, end='')

output
```
